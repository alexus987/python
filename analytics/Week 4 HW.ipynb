```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
### Моё любимое непрерывное распределение - нормальное ###

# зададим нормально распределенную случайную величину
norm_rv = sts.norm(0, 1)

# сгенерируем 1000 значений
sample = norm_rv.rvs(1000)
```

```python
from statsmodels.distributions.empirical_distribution import ECDF
ecdf = ECDF(sample)
plt.step(ecdf.x, ecdf.y, label='MY CDF')

#теоретическая плотность распределения
x = np.linspace(-4,4,100)
#Следующая функция возвращает значение функции распределения нормальной случайной величины в точке, соответствующей её аргументу X
cdf = norm_rv.cdf(x)
plt.plot(x, cdf, label='theoretical CDF')

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper left')
```

```python
poisson_rv = sts.poisson(5)
poisson_rv.rvs(1000) 

x = np.linspace(0,30,100)
for l in [5,10,50]:
    rv = sts.poisson(l)
    cdf = rv.cdf(x)
    plt.step(x, cdf, label="$\lambda=%s$" % l)
plt.legend()
plt.title("CDF (poisson)")

plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```
